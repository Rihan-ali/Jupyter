What is Data Cleaning ?
-


Data Cleaning:- Data cleaing is the process of Identify  and correct the incorrectness ,incompleteness ,inaccuracies and error in the raw  to 
inprove the data quality and consistancy.And ensure that data is ready for Analysis.
In data cleaning we handle the null values and duplicates record as well.

```python
#Change the Column name
df.rename(columns={'Unnamed: 0' : 'Serial_Number'},inplace=True)
```

```python
#Extrac the firsty wait value using loop and indexing
for i in range(len(df)):
    df['Weight']=df['Weight'][i][0 :4]
```

```python
#change the data type
df['Weight']=df['Weight'].astype(float)
```

```python
#Replace '?' unnacessary character with 0 and  then chanage the  data type and  and then replace 0 with mean of the total to maintain the byseness of data.
df['Inches']=df['Inches'].replace('?',0)

df['Inches']=df['Inches'].astype(float)

```

```python
df['Inches']=df['Inches'].replace(0,df['Inches'].mean(),inplace=True)
```
